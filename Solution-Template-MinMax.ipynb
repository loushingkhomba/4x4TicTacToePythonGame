```python
class minMaxAlgo():
    def __init__(self):
        self.initialize_board()

    def initialize_board(self):
        self.current_state = [['-','-','-','-'],
                            ['-','-','-','-'],
                            ['-','-','-','-'],
                            ['-','-','-','-']]

        # Player X always plays first
        self.player_turn = 'X'

    def draw_board(self):
        for i in range(0, 4):
            for j in range(0, 4):
                print('{}|'.format(self.current_state[i][j]), end=" ")
            print()
        print()

    # Determines if the made move is a legal move
    def is_validmove(self, px, py):
        if px < 0 or px > 3 or py < 0 or py > 3:
            return False
        elif self.current_state[px][py] != '-':
            return False
        else:
            return True
    
    def play_alpha_beta_theta(self):
        while True:
            self.draw_board()
            self.result = staticEval.is_endofgame(self)
                
            if self.result != None:
                if self.result == 'X':
                    print('The winner is X!')
                elif self.result == 'O':
                    print('The winner is O!')
                elif self.result == 'T':
                    print('The winner is T!')
                elif self.result == '-':
                    print("It's a tie!")


                self.initialize_board()
                return

            if self.player_turn == 'X':

                while True:
                    
                    (m, qx, qy) = alphaBetaPruning.min_alpha_beta(self,-3, 3)
                    
                    print('Recommended move: X = {}, Y = {}'.format(qx, qy))

                    px = int(input('Insert the X coordinate: '))
                    py = int(input('Insert the Y coordinate: '))

                    qx = px
                    qy = py

                    if self.is_validmove(px, py):
                        self.current_state[px][py] = 'X'
                        self.player_turn = 'O'
                        break
                    else:
                        print('The move is not valid! Try again.')

            elif self.player_turn == 'O':
                (m, px, py) = alphaBetaPruning.max_alpha_beta(self,-3, 3)
                print("x-",px,"y-",py)
                self.current_state[px][py] = 'O'
                self.player_turn = 'T'
            elif self.player_turn == 'T':
                (m, tx, ty) = alphaBetaPruning.max_alpha_theta(self,-3, 3)
                print("x-",tx,"y-",ty)
                self.current_state[tx][ty] = 'T'
                self.player_turn = 'X'
```

### Choice and implementation of the Static Evaluation Function.

```python
class staticEval(): 
    # Checks if the game has ended and returns the winner in each case
    def is_endofgame(self):
    # Check for column-wise win
        for i in range(0, 4):
            if (self.current_state[0][i] != '-' and
                self.current_state[0][i] == self.current_state[1][i] and
                self.current_state[0][i] == self.current_state[2][i]):
                return self.current_state[0][i]
            elif (self.current_state[3][i] != '-' and
                self.current_state[3][i] == self.current_state[2][i] and
                self.current_state[3][i] == self.current_state[1][i]):
                return self.current_state[3][i]

        def equalThree(a,b,c):
            if (a=='X'):
                if (a==b and b==c):
                    return True
            elif(a=='O'):
                if (a==b and b==c):
                    return True
            elif(a=='T'):
                if (a==b and b==c):
                    return True
    # Check for row-wise win
        for i in range(0, 4):
            if (self.current_state[i] == ['X', 'X', 'X', '-'] or self.current_state[i] == ['-', 'X', 'X', 'X'] or 
                self.current_state[i] == ['X', 'X', 'X', 'T'] or self.current_state[i] == ['T', 'X', 'X', 'X'] or 
                self.current_state[i] == ['X', 'X', 'X', 'O'] or self.current_state[i] == ['O', 'X', 'X', 'X'] or 
                self.current_state[i] == ['X', 'X', 'X', 'X']):
                return 'X'
            elif (self.current_state[i] == ['O', 'O', 'O', '-'] or self.current_state[i] == ['-', 'O', 'O', 'O'] or
                self.current_state[i] == ['O', 'O', 'O', 'X'] or self.current_state[i] == ['X', 'O', 'O', 'O'] or
                self.current_state[i] == ['O', 'O', 'O', 'T'] or self.current_state[i] == ['T', 'O', 'O', 'O'] or
                self.current_state[i] == ['O', 'O', 'O', 'O']):
                return 'O'
            elif (self.current_state[i] == ['T', 'T', 'T', '-'] or self.current_state[i] == ['-', 'T', 'T', 'T'] or
                self.current_state[i] == ['T', 'T', 'T', 'X'] or self.current_state[i] == ['X', 'T', 'T', 'T'] or
                self.current_state[i] == ['T', 'T', 'T', 'O'] or self.current_state[i] == ['O', 'T', 'T', 'T'] or
                self.current_state[i] == ['T', 'T', 'T', 'T']):
                return 'T'

    # Check for win over the 1st diagonal 
        if (self.current_state[0][0] != '-' and
            self.current_state[0][0] == self.current_state[1][1] and
            self.current_state[0][0] == self.current_state[2][2]):
            return self.current_state[0][0]
    # Check for win over the 2nd diagonal 
        if (self.current_state[3][3] != '-' and
            self.current_state[3][3] == self.current_state[2][2] and
            self.current_state[3][3] == self.current_state[1][1]):
            return self.current_state[3][3]
    # Check for win over the 3rd diagonal 
        if (self.current_state[0][1] != '-' and
            self.current_state[0][1] == self.current_state[1][2] and
            self.current_state[0][1] == self.current_state[2][3]):
            return self.current_state[0][1]
    # Check for win over the 4th diagonal 
        if (self.current_state[1][0] != '-' and
            self.current_state[1][0] == self.current_state[2][1] and
            self.current_state[1][0] == self.current_state[3][2]):
            return self.current_state[1][0]

    # Check for win over the 5th diagonal 
        if (self.current_state[0][3] != '-' and
            self.current_state[0][3] == self.current_state[1][2] and
            self.current_state[0][3] == self.current_state[2][1]):
            return self.current_state[0][2]
    # Check for win over the 6th diagonal 
        if (self.current_state[3][0] != '-' and
            self.current_state[3][0] == self.current_state[2][1] and
            self.current_state[3][0] == self.current_state[1][2]):
            return self.current_state[3][0]
    # Check for win over the 7th diagonal 
        if (self.current_state[0][2] != '-' and
            self.current_state[0][2] == self.current_state[1][1] and
            self.current_state[0][2] == self.current_state[2][0]):
            return self.current_state[0][2]
    # Check for win over the 8th diagonal 
        if (self.current_state[1][3] != '-' and
            self.current_state[1][3] == self.current_state[2][2] and
            self.current_state[1][3] == self.current_state[3][1]):
            return self.current_state[1][3]

    # Is whole board full?
        for i in range(0, 4):
            for j in range(0, 4):
            # There's an empty field, we continue the game
                if (self.current_state[i][j] == '-'):
                    return None

    # It's a tie!
        return '-'
```
